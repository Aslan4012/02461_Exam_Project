```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def confidence_interval(data):
    mean = np.mean(data)
    std = np.std(data, ddof=1)
    n = len(data)
    z = 1.96
    lower = mean - z * (std/np.sqrt(n))
    upper = mean + z * (std/np.sqrt(n))
    return lower, upper

def sample_size(data, margin_error):
    std = np.std(data,ddof=1)
    z = 1.96
    n = (z * std / margin_error) ** 2
    return n
```
